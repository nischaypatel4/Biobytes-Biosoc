Read the following abstract and answer the questions

The newest house in your data isn't that new. A few potential explanations for this:

 -They haven't built new houses where this data was collected.

 -The data was collected a long time ago. Houses built after the data publication wouldn't show up.

If the reason is explanation #1 above, does that affect your trust in the model you build with this data? What about if it is reason #2?

Answer:    If the reason is #1, then it does not affect the accuracy of the model, because it can be considered as the "latest" data

If the reason is #2, then it can hamper the accuracy of the model becuase the data is not up to date and if the model is used to give a prediction on newer data, there is a high chance that it might fail to predict the correct price.


How could you dig into the data to see which explanation is more plausible?

Answer: We figure out the average difference within YearBuilt, that is, how frequent a house is being built(as shown in the next code cell).
By looking at the value that is approximately 1.23, that is, on an average, there is a house being built every 1-2 years.

Now note that the latest house that has been built is 2010. The year going on is 2024, that is, it has been 24 years since a house has been built which according the above found metric is a long time, therefore, it can be considered that the dataset provided belongs to the second category.

```python
sample = dataset['YearBuilt']
sample = list(sorted(sample))
sample = list(set(sample))
print("Average difference in YearBuilt in the area:", (sample[-1]-sample[0])/len(sample))
print("Latest House built in the area:", sample[-1])
```

```python
Y = pd.DataFrame(dataset['SalePrice'])
```

```python
X = dataset[['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']]
```

```python
NewData = X.join(Y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
LinearModel = LinearRegression()
DTModel = DecisionTreeRegressor()
```

```python
LinearModel.fit(X, Y)
DTModel.fit(X, Y)
```

```python
LinearModel.predict(X.head())
```

```python
Y.head()
```

```python
DTModel.predict(X.head())
```

**Anything Surprising?**

It is surprising to see that the DecisionTreeRegressor is performing extraordinarily well and the prediction are very accurate, while on the other hand the Linear Regression model is not able to predict very well even in case of in-sample testing.

```python
from sklearn.metrics import mean_absolute_error as mae
```

```python
linear_prediction = LinearModel.predict(X)
DT_prediction = DTModel.predict(X)
print("LR Error:", mae(Y, linear_prediction))
print("DTR Error:", mae(Y, DT_prediction))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=1, test_size=0.2)
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
newDTModel = DecisionTreeRegressor(random_state=1)
newDTModel.fit(X_train, Y_train)
```

```python
print("Actual Values:",list(Y_test.head()['SalePrice']),"\nPredictions:", newDTModel.predict(X_test.head()))
```

**What do you notice that is different from what you saw with in-sample predictions (which are printed after the top code cell in this page)?**

Here, we can see that the predictions vary a lot from the actual values unlike the in-sample testing done previously wherein the prediction were very accurate!

**Do you remember why validation predictions differ from in-sample (or training) predictions?**

In case of in-sample training, the model is in a way remembering the outputs of specific inputs and as a result it was giving very accurate predictions but in the second case, it wasn't very good at predicting the values. This is implies that there can be a case of overfitting.

```python
mae(Y_test, newDTModel.predict(X_test))
```

```python
LinearModel = LinearRegression()
LinearModel.fit(X_train, Y_train)
mae(Y_test, LinearModel.predict(X_test))
```

**Is that MAE good? There isn't a general rule for what values are good that applies across applications.**

In comparision to the linear regression model trained and tested over the same data, the mae is even worse which is a complete opposite of the in-sample case.
