```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
```

```python
data = pd.read_csv('/content/dataset.csv')
```

```python
data.head(8)
```

```python
stats = data.describe()
stats
```

```python
avg_LotArea = stats['LotArea']['mean']
max_SalePrice = stats['SalePrice']['max']
mean_SalePrice = stats['SalePrice']['mean']
std_SalePrice = stats['SalePrice']['std']
print(avg_LotArea)
print(max_SalePrice)
```

If they have not build new houses where the data was collected, it does affect my trust in the model in the sense that my model will only be applicable to this specific area now. If the area was to affect the prices of the house then my model would not be able to predict correctly the prices of a house built in a new area.

If it has been a while since the data has been collected, my trust in the model will not have deteriorated much (excluding inflation which can be accounted for)
A reason for mistrust could be that time has introduced new factors which have now affected the prices of all houses in the dataset, example some schools open in a particular area only and increase the price of those particular houses.

We can check the distribution of the location of the houses versus the age of the houses.
If the new houses are located at distances further from the older houses it means that not many new houses are being built in the primary area of interest.
If the new houses are located at distances roughly matching the distances of the older houses then it means houses are still being built in this area and the limiting factor is the age of the dataset.

```python
data.dropna(axis=0)
data.columns.isnull()
y = data['SalePrice']
data.shape
```

```python
x = data[['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']]
print(x.shape)
print(y.shape)
```

```python
m1 = LinearRegression()
m1.fit(x,y)
m2 = DecisionTreeRegressor()
m2.fit(x,y)

print("Making predictions for the following 5 houses:")
print(x.head(5))
print("Linear Regression:")
print(m1.predict(x.head(5)))
print("Decision Tree Regressor:")
print(m2.predict(x.head(5)))
print("Actual Values:")
print(y.head(5))
```

```python
y1 = m1.predict(x)
y2 = m2.predict(x)
e1 = mean_absolute_error(y,y1)
e2 = mean_absolute_error(y,y2)

print(e1)
print(e2)
```

Surprising is the fact that the Decision Tree correctly predicted all values

```python
t_x,v_x,t_y,v_y = train_test_split(x,y,random_state=1)
```

```python
model = DecisionTreeRegressor(random_state=1)
model.fit(t_x,t_y)
print("Predicted prices on validation data:")
print(model.predict(v_x.head()))
print("Actual values:")
print(v_y.head())
```

This time the predicted values for the houses outside of the training sample are not as accurate as the last time where the samples were from the training data itself.


```python
print(mean_absolute_error(v_y,model.predict(v_x)))
print(mean_SalePrice)
print(std_SalePrice)
```

Considering that the predicted sale price lies generally within nearly three-eights of a standard deviation of the mean sale price the model is fairly decent.
