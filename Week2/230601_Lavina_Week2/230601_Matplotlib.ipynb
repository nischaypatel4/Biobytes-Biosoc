## Exercise 1

**Follow along with these steps**
* Create a figure object called fig using plt.figure()
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
# Data
x = [0,20,40,60,80,100]
y = [0,40,80,120,160,200]

# Create a figure object
fig = plt.figure()

# Add axes to the figure canvas
ax = fig.add_axes([0, 0, 1, 1])

# Plot data on the axes
ax.plot(x, y)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Title')
```

**Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
# Create a figure object
fig = plt.figure()

# Add axes to the figure canvas
ax1 = fig.add_axes([0, 0, 1, 1])       # Main axes
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])  # Inset axes

# Show the plot
plt.show()
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python


# Updated data ranges
x_main = np.linspace(0, 100, 1000)  # Range from 0 to 100

z_main = np.linspace(0, 100, 1000)*np.linspace(0, 100, 1000)

x_inset = np.linspace(20, 22, 200)  # Range from 20 to 22
y_inset = np.linspace(40, 44, 200)  # Range from 40 to 44

# Creating the main figure and axes
fig, ax1 = plt.subplots()
ax1.plot(x_main, z_main, 'b-')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

# Creating the inset axes
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])  # [left, bottom, width, height]
ax2.plot(x_inset, y_inset, 'b-')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')

# Setting limits for the inset plot
ax2.set_xlim(20, 22)
ax2.set_ylim(30, 55)

plt.show()
```

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
# Updated data ranges
x_main = np.linspace(0, 100, 1000)  # Range from 0 to 100
y_main = np.linspace(0, 100, 1000) * 2
z_main = np.linspace(0, 100, 1000) * np.linspace(0, 100, 1000)

# Creating the main figure and axes
fig, ax1 = plt.subplots(nrows=1, ncols=2)
ax1[0].plot(x_main, y_main, 'b--', label='y', linewidth=2)
ax1[1].plot(x_main, z_main, 'r-', label='z', linewidth=1.5)
ax1[0].set_xlabel('X')
ax1[1].set_xlabel('X')
ax1[0].set_ylabel('Y')
ax1[1].set_ylabel('Z')
plt.show()
```
