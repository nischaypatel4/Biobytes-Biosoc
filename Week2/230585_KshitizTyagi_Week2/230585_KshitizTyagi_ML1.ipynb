```python
targ_vary=df['SalePrice']
targ_vary.head()
```

```python
#7
feature_x=df[['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']]
feature_x.head()
```

```python
#8
df.columns.isnull()
```

```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#9
dflin=LinearRegression()
dflin.fit(feature_x,targ_vary)
```

```python
print(dflin.predict(feature_x.head()))
targ_vary.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
dfdis=DecisionTreeRegressor()
dfdis.fit(feature_x,targ_vary)
```

```python
print(dfdis.predict(feature_x.head()))
targ_vary.head()
```

The suprising this is that decision tree fits to the data very accurately while regression is not a perfect fit

```python
#10 MAE for linear
from sklearn.metrics import mean_absolute_error
prelin=dflin.predict(feature_x)
mean_absolute_error(targ_vary,prelin)
```

```python
#MAE for decision
predis=dfdis.predict(feature_x)
mean_absolute_error(targ_vary,predis)
```

```python
#11
from sklearn.model_selection import train_test_split
train_x, val_x, train_y, val_y = train_test_split(feature_x, targ_vary, random_state = 1)
```

```python
#12
dftrdis=DecisionTreeRegressor()
dftrdis.fit(train_x,train_y)
```

```python
#13
print(dftrdis.predict(val_x.head()))
val_y.head()
```

14- when we were making prediction over the training data the data fits very well but when we seperated the testing and training data it was found that this model is not of great fit to the data as well

```python
#15
mean_absolute_error(val_y,dftrdis.predict(val_x))
```

```python
#16 To check whether it is good or not lets compare the same with linear regression
dftrlin=LinearRegression()
dftrlin.fit(train_x,train_y)
mean_absolute_error(val_y,dftrlin.predict(val_x))
```

16 so on comparing the same with linear regression the MAE of decision tree is more. So this MAE is not good on comparing.
