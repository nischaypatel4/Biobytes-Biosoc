## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python



x = np.linspace(0, 100, 6)
y = x*2
z = x**2


fig, axs = plt.subplots(nrows=1, ncols=2)


axs[0].plot(x, y, linewidth=2, linestyle='--', color='blue')
axs[0].set_title('(x, y)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')


axs[1].plot(x, z, color='red')
axs[1].set_title('(x, z)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('z')


plt.tight_layout()


plt.show()
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
 x= np.linspace(0, 100, 6)
y = x*2
z = x**2


fig, axs = plt.subplots(nrows=1, ncols=2,figsize=(12, 3))


axs[0].plot(x, y , color='blue')
axs[0].set_title('(x, y)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')


axs[1].plot(x, z, color='red',linestyle='--',linewidth=2,)
axs[1].set_title('(x, z)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('z')


plt.tight_layout()


plt.show()
```
