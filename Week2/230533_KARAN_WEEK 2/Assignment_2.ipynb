Explanation #1: No New Houses Built:-
1.If the housing market in the area hasn't changed much over time, the factors that affect house prices probably haven't changed either. So, the model built with the existing data might still work well.
2.The model will be good at predicting prices for older houses or houses built around the same time. If you're mainly interested in these types of houses, you can trust the model.
Explanation #2: Data Collected a Long Time Ago:-
1.The housing market can change a lot over time because of things like the economy, interest rates, population growth, and new buildings. A model using old data might not reflect these current changes.
2.Changes in housing rules, zoning laws, or taxes can greatly impact the housing market, making older data less useful.


How could you dig into the data to see which explanation is more plausible?
ans:-1.Look for metadata or documentation that indicates when the data was collected. This can help you determine if the dataset is outdated.
2.Cross-Reference Recent Sales Data: Compare your data with recent home sales to spot any missing new houses.
3.Look at recent building permits in the area to see if new houses have been built.

```python
melb_data.columns
```

```python
y = melb_data[['SalePrice']]
y
```

```python
X=melb_data[['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']]
X
```

```python
new_dataset = pd.concat([X, y], axis=1)
print(new_dataset.head())
```

```python
new_dataset.columns.isnull()
#no missing value
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


x = new_dataset.drop(columns = ['SalePrice'])
x = x.select_dtypes(include=['float64', 'int64'])
x.head()

melbourne_model1 = LinearRegression()
melbourne_model1.fit(x, y)



print("Making predictions for the following 5 houses:")
print(x.head())
print("The predictions are")
print(melbourne_model1.predict(x.head()))
```

```python
from sklearn.metrics import mean_absolute_error

predicted_prices_LinearReg = melbourne_model1.predict(x)

mean_absolute_error(y, predicted_prices_LinearReg)
```

```python
from sklearn.tree import DecisionTreeRegressor

melbourne_model2 = DecisionTreeRegressor(random_state=2)
melbourne_model2.fit(x, y)

print("Making predictions for the following 5 houses:")
print(x.head())
print("The predictions are")
print(melbourne_model2.predict(x.head()))
```

```python
from sklearn.metrics import mean_absolute_error

predicted_prices_DecisionTree= melbourne_model2.predict(x)

mean_absolute_error(y, predicted_prices_DecisionTree)
```

```python
from sklearn.model_selection import train_test_split
train_x, val_x, train_y, val_y = train_test_split(x, y, random_state = 0)

train_x.head()

val_x.head()

melbourne_model2 = DecisionTreeRegressor()

melbourne_model2.fit(train_x, train_y)

val_predictions = melbourne_model2.predict(val_x)
print(mean_absolute_error(val_y, val_predictions))
```
