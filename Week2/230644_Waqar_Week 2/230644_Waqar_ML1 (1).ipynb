```python
y = dataset['SalePrice']
x = dataset[['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']]
df = pd.concat([x, y], axis=1)
df.head()
```

```python
missing = df.isnull().sum()
missing
```

```python
df = df.dropna(axis=1, how='any')
print(df.isnull().sum())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
x = df.drop(columns=['SalePrice'])
y = df['SalePrice']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

lr = LinearRegression()
lr.fit(x_train, y_train)
lr_train_predictions = lr.predict(x_train)
lr_predictions = lr.predict(x_test)

dt = DecisionTreeRegressor( random_state=1)
dt.fit(x_train, y_train)
dt_train_predictions = dt.predict(x_train)
dt_predictions = dt.predict(x_test)

comparison_train_df = pd.DataFrame({'Actual': y_train,'LR': lr_train_predictions,'DT': dt_train_predictions})
comparison_train_df.head()
```

the decision tree predictions are 100% accurate.

```python
from sklearn.metrics import mean_absolute_error
lr_train_mae = mean_absolute_error(y_train, lr_train_predictions)
dt_train_mae = mean_absolute_error(y_train, dt_train_predictions)
print(lr_train_mae)
print(dt_train_mae)
```
