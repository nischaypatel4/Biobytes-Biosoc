```python
y = df.SalePrice
y
```

```python
x = df[['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']]
x
```

```python
print(x.isnull().any())
print('')
print(y.isnull().any())
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
lr_model = LinearRegression()
dt_model = DecisionTreeRegressor()

lr_model.fit(x,y)
dt_model.fit(x,y)
```

```python
print('Prediction using LR')
print(lr_model.predict(x.head()))
print('Prediction using DT')
print(dt_model.predict(x.head()))
print('Actual Values')
print(y.head())
```

```python
from sklearn.metrics import mean_absolute_error

predicted_lr = lr_model.predict(x)
mean_absolute_error(y,predicted_lr)
```

```python
predicted_dt = dt_model.predict(x)
mean_absolute_error(y,predicted_dt)
```

```python
from sklearn.model_selection import train_test_split
train_x, val_x, train_y, val_y = train_test_split(x, y, random_state=1)
```

```python
dt_model.fit(train_x,train_y)
print(dt_model.predict(val_x.head()))
print(val_y.head())
```

In-sample predictions and actual values were coming out to be equal whereas here there is a difference in the two.

The two differ because the in-sample set is used to train the model whereas the validation set is used to actually assess and test the model(which may contain inaccuracies).

```python
mean_absolute_error(val_y,dt_model.predict(val_x))
```
