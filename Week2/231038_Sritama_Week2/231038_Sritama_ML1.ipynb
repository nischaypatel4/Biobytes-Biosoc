The newest house in your data isn't that new. A few potential explanations for this:
 -They haven't built new houses where this data was collected. (1)
 -The data was collected a long time ago. Houses built after the data publication wouldn't show up. (2)

I. In case of the first reason, there may or may not be reconsiderations about the current dataset. It might prove to be just as reliable or there might be new factors that come into play such as climate, inflation, etc.

II. In the second case, it definitely affects our current dataframe. Recent data is given much higher importance and future data is also more likely to be more similar to fresh data in comparison to data collected ten years ago.

```python
build_years = dataset["YearBuilt"]
bins = build_years.max() - build_years.min()

build_years.hist(bins=bins)
```
