```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset.csv')
```

```python
df.head(8)
```

```python
df.describe()
```

```python
Average_Lot_Area = round(np.mean(df['LotArea']))
Max_SalePrice = round(np.max(df['LotArea']))
Average_Lot_Area, Max_SalePrice
```

```python
y = np.array(df[['SalePrice']])
X = df[['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']]
X
```

```python
X.info()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

LinearRegressionModel = LinearRegression()
LinearRegressionModel.fit(X, y)
predictionLinearRegression = LinearRegressionModel.predict(X)

DecisionTreeModel = DecisionTreeRegressor()
DecisionTreeModel.fit(X, y)
predictionDecisionTree = DecisionTreeModel.predict(X)
prediction_df = pd.DataFrame({'Actual': y.flatten(),
                              'Linear_Pred': predictionLinearRegression.flatten(),
                              'Tree_Pred': predictionDecisionTree.flatten()})
prediction_df.head()
```

The Decision Tree Model fits too perfectly to the training, in ML Lingo
We say that the model is overfit.

```python
from sklearn.metrics import mean_absolute_error
mae_LinearRegressionModel = mean_absolute_error(y, predictionLinearRegression)
mae_DecisionTreeModel = mean_absolute_error(y, predictionDecisionTree)
(mae_LinearRegressionModel, mae_DecisionTreeModel)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
tree_model = DecisionTreeRegressor(random_state=1)
tree_model.fit(X_train, y_train)
y_valid_pred = tree_model.predict(X_valid)
```

```python
validation_comparison_df = pd.DataFrame({'Actual': y_valid.flatten(),
                                         'Tree_Pred': y_valid_pred.flatten()})
validation_comparison_df.head()
```

Clearly, the model performs poorly on the test dataset and performs extremely well on the training dataset. So, this model is overfit and has high variance.

```python
mae_valid_tree = mean_absolute_error(y_valid, y_valid_pred)
mae_valid_tree
```
