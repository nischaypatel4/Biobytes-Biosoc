```python
y = data.loc[:, ['SalePrice']]
X = data.loc[:, ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']]
new_data = pd.concat([X, y], axis=1)
```

```python
new_data.isnull().sum()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
model1 = LinearRegression()
model2 = DecisionTreeClassifier()
```

```python
model1.fit(X,y)
```

```python
model2.fit(X,y)
```

```python
new_data.head(8)
```

```python
lin_reg_pred = pd.DataFrame(model1.predict(X))
lin_reg_pred.head(8)
```

```python
dec_tree_pred = pd.DataFrame(model2.predict(X))
dec_tree_pred.head(8)
```

#### Yeah the linear regression model predicted somewhat deviated values but decission tree model predicted exact value ###
#### it is because linear regression fits a single line to the training data, and the predictions are on this line, so it is doing underfitting while decission tree is doing overfitting #####

```python
from sklearn.metrics import mean_absolute_error
mae_lin_reg = mean_absolute_error(y, lin_reg_pred)
mae_dec_tree = mean_absolute_error(y, dec_tree_pred)
```

```python
print(mae_lin_reg)
print(mae_dec_tree)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=1)
model2.fit(X_train, y_train)
```

```python
dec_tree_pred2 = pd.DataFrame(model2.predict(X_test))
dec_tree_pred2
```

```python
y_test
```

```python
mae_dec_tree2 = mean_absolute_error(y_test, dec_tree_pred2)
print(mae_dec_tree2)
```

#### Observation: mae increased drastically. It is because earlier we were predicting the data which the model has already seen so it was doing overfitting. But now we are predicting on a new data, so the chances of error has increased. ####

#### there isn't any general rule whether the MAE is good or not, it depends on the context for which we are training the model. In our context the MAE value is pretty lesser tham the exact outcome values, so our model is satisfactory though it is not very good. ####
