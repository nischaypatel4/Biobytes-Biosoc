# Logistic Regression

Now it's time to do a train test split, and train our model!

You'll have the freedom here to choose columns that you want to train on!

**Split the data into training set and testing set using train_test_split**

```python
from sklearn.model_selection import train_test_split
X = ad_data.select_dtypes(exclude='object')
Y = ad_data['Clicked on Ad']
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, random_state = 0)
```

```python
train_X = train_X.values
test_X = test_X.values
train_Y = train_Y.values
test_Y = test_Y.values
```

```python
train_X = train_X.T
train_Y = train_Y.reshape(1, train_X.shape[1])

test_X = test_X.T
test_Y = test_Y.reshape(1, test_X.shape[1])
```

**Train and fit a logistic regression model on the training set.**

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))

def model(X, Y, learning_rate, iterations):

    m = X.shape[1]
    n = X.shape[0]

    W = np.zeros((n, 1))
    B = 0

    cost_list = []
    epsilon = 1e-8

    for i in range(iterations):

        Z = np.dot(W.T, X) + B
        A = sigmoid(Z)

        A = np.clip(A, epsilon, 1 - epsilon)

        cost = -(1/m) * np.sum(Y * np.log(A + epsilon) + (1 - Y) * np.log(1 - A + epsilon))

        dW = (1/m) * np.dot(X, (A - Y).T)
        dB = (1/m) * np.sum(A - Y)

        W = W - learning_rate * dW
        B = B - learning_rate * dB

        cost_list.append(cost)

        if(i%(iterations/10) == 0):
            print("cost after ", i, "iteration is : ", cost)

    return W, B, cost_list
```

```python
iterations = 100000
learning_rate = 0.0015
W, B, cost_list = model(train_X, train_Y, learning_rate=learning_rate, iterations=iterations)
```
